```python
# Importing Require libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading CSV file 
covid_confirm =pd.read_csv("time_series_covid19_confirmed_global.csv")
covid_death = pd.read_csv("time_series_covid19_deaths_global.csv")
covid_recovered = pd.read_csv("time_series_covid19_recovered_global.csv")
```

```python
# Checking shape of files
print(f" Raw confirm case data shape: {covid_confirm.shape}")
print(f" Raw death case data shape: {covid_death.shape}")
print(f" Raw recovered case data shape: {covid_recovered.shape}")
```

```python
# Unpivot a DataFrame
## In file Dates are in columns it easy to analysis when dates in row
covid_confirm_case= pd.melt(covid_confirm , id_vars=("Province/State","Country/Region","Lat","Long"),var_name="Dates")

covid_death_case= pd.melt(covid_death , id_vars=("Province/State","Country/Region","Lat","Long"),var_name="Dates")

covid_recovered_case= pd.melt(covid_recovered , id_vars=("Province/State","Country/Region","Lat","Long"),var_name="Dates")

print(f" Raw confirm case data shape: {covid_confirm_case.shape}")
print(f" Raw death case data shape: {covid_death_case.shape}")
print(f" Raw recovered case data shape: {covid_recovered_case.shape}")
```

```python
# filling Null values
##Removing Null Values on column Province/State by using Country/Region
covid_confirm_case["Province/State"].fillna(covid_confirm_case["Country/Region"], inplace=True)
covid_death_case["Province/State"].fillna(covid_death_case["Country/Region"], inplace=True)
covid_recovered_case["Province/State"].fillna(covid_recovered_case["Country/Region"], inplace=True)


covid_confirm_case.head(20)
```

```python
print(f" Raw confirm case data shape: {covid_confirm_case.shape}")
print(f" Raw death case data shape: {covid_death_case.shape}")
print(f" Raw recovered case data shape: {covid_recovered_case.shape}")
```

```python
## Recheck of null value in all Columns
covid_confirm_case.isnull().sum()
```

```python
# Here seen that value shown in column " counts of cases " i.e confirmed case on date
covid_confirm_case.columns=covid_confirm_case.columns.str.replace("value","Confirmed_Case")
covid_death_case.columns=covid_death_case.columns.str.replace("value","Death_Case")
covid_recovered_case.columns=covid_recovered_case.columns.str.replace("value","Recovered_Case")
```

```python
# Change Date column object datatype to Datetime datatype
from datetime import datetime
covid_confirm_case["Dates"]=pd.to_datetime(covid_confirm_case["Dates"])
covid_death_case["Dates"]=pd.to_datetime(covid_death_case["Dates"])
covid_recovered_case["Dates"]=pd.to_datetime(covid_recovered_case["Dates"])

covid_confirm_case.info()
```

```python
# Joining all three file 
final_data= covid_confirm_case.merge(covid_death_case[['Province/State','Country/Region','Dates','Death_Case']],
                                     how = 'left', 
                                     left_on = ['Province/State','Country/Region','Dates'], 
                                     right_on = ['Province/State', 'Country/Region','Dates'])

final_data
```

```python
final_data2= final_data.merge(covid_recovered_case[['Province/State','Country/Region','Dates','Recovered_Case']],
                                     how = 'left', 
                                     left_on = ['Province/State','Country/Region','Dates'], 
                                     right_on = ['Province/State', 'Country/Region','Dates'])

final_data2
```

```python
# Extrating Month and year from Dates
final_data2["Month-Year"]= final_data2["Dates"].dt.strftime("%b-%Y")
final_data2.head()
```

```python
# Doing feature engineering for Daily Confirmed Cases , Daliy Recovered Cases and Daily Deaths Cases

final_data3 = final_data2.copy()

final_data3["Date+1"] = final_data3['Dates']+pd.Timedelta(days=1)
final_data3.rename(columns={'Confirmed_Case':'confirmdate+1',
                            'Death_Case':'Deathdate+1',
                            'Recovered_Case':'Recoverdate+1','Dates':'Date-1'}, inplace=True)
final_data3.head()
```

```python
final_data4=final_data2.merge(final_data3[['Province/State','Country/Region','confirmdate+1','Deathdate+1','Recoverdate+1','Date+1','Date-1']],
                             how='left',
                             left_on=['Province/State','Country/Region','Dates'],
                             right_on=['Province/State','Country/Region','Date+1'])
final_data4.head()
```

```python
final_data4["Confirm_Daily"]=final_data4["Confirmed_Case"]-final_data4["confirmdate+1"]
final_data4["Death_Daily"]=final_data4["Death_Case"]-final_data4["Deathdate+1"]
final_data4["Recovered_Daily"]=final_data4["Recovered_Case"]-final_data4["Recoverdate+1"]

final_data4.head()
```

```python
## First day daily case are equal to confirmed case 
final_data4["Confirm_Daily"].loc[final_data4['Dates']=='2020-01-22']=final_data4["Confirmed_Case"]
final_data4["Death_Daily"].loc[final_data4['Dates']=='2020-01-22']=final_data4["Death_Case"]
final_data4["Recovered_Daily"].loc[final_data4['Dates']=='2020-01-22']=final_data4["Recovered_Case"]

# Removing columns
del final_data4['confirmdate+1']
del final_data4['Deathdate+1']
del final_data4['Recoverdate+1']
del final_data4['Date+1']
del final_data4['Date-1']

final_data4.head()
```

```python

final_data4["Confirm_Daily"]=np.where(final_data4["Confirm_Daily"]<0,0,final_data4["Confirm_Daily"])
final_data4["Death_Daily"]=np.where(final_data4["Death_Daily"]<0,0,final_data4["Death_Daily"])
final_data4["Recovered_Daily"]=np.where(final_data4["Recovered_Daily"]<0,0,final_data4["Recovered_Daily"])
```
